step 3: Data Cleaning and Transformation

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the training data
train_data = pd.read_csv('../data/train_data.csv')

# Set timestamp as index
train_data['Timestamp'] = pd.to_datetime(train_data['Timestamp'])
train_data.set_index('Timestamp', inplace=True)

# Aggregate data to weekly consumption
trainweekly_data = train_data.resample('W').sum()
trainweekly_data.to_csv('../data/weekly_data_train.csv')


# Load the testing data
test_data = pd.read_csv('../data/test_data.csv')

# Set timestamp as an index
test_data['Timestamp'] = pd.to_datetime(test_data['Timestamp'])
test_data.set_index('Timestamp', inplace=True)

# Aggregate data to weekly consumption
testweekly_data = train_data.resample('W').sum()
testweekly_data.to_csv('../data/weekly_data_test.csv')

print("Transformation completed")

```

```python
# Plot the weekly consumption data
plt.figure(figsize=(12, 6))
plt.plot(trainweekly_data.index, trainweekly_data['Amount'], marker='o')
plt.title('Weekly Consumption Amount')
plt.xlabel('Date')
plt.ylabel('Amount ($)')
plt.grid(True)
plt.show()
```
